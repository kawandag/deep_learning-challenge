## Preprocessing

```python

# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import requests
import io

#  Import and read the charity_data.csv.
import pandas as pd
url = "https://raw.githubusercontent.com/kawandag/deep_learning-challenge/main/charity_data.csv?token=GHSAT0AAAAAABNOS74ZZNTPAZB6RDFLEGNYYRBLNWQ"
s = requests.get(url).content



```

```python
application_df = pd.read_csv(io.StringIO(s.decode('utf-8')),low_memory=False)
```

```python
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
new_app = application_df.drop(columns={'EIN', 'NAME'})
new_app
```

```python
# Determine the number of unique values in each column.
new_app.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_df = new_app["APPLICATION_TYPE"].value_counts()
app_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
#cutoff_value = 500
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_df[app_df < 500].index)

# Replace in dataframe
for app in application_types_to_replace:
    new_app['APPLICATION_TYPE'] = new_app['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
new_app['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_df = new_app["CLASSIFICATION"].value_counts()
class_df
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_df[class_df >1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
#cutoff value = 100
# use the variable name `classifications_to_replace`
classifications_to_replace =  list(class_df[class_df < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    new_app['CLASSIFICATION'] = new_app['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
new_app['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(new_app)
app_dummies
```

```python
# Split our preprocessed data into our features and target arrays
X = app_dummies.drop(["IS_SUCCESSFUL"], axis='columns').values
y = app_dummies["IS_SUCCESSFUL"].values


# Split the preprocessed data into a training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(X,y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(x_train)

# Scale the data
X_train_scaled = X_scaler.transform(x_train)
X_test_scaled = X_scaler.transform(x_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_feat = len(x_train[0])
layer1 = 100
layer2 = 40

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=layer1, input_dim=input_feat, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=layer2, input_dim=layer1, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, input_dim=layer2, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
